## Summary

```python
import pandas as pd

#Loading the dataset
df = pd.read_csv('../Raw_Data/PRODUCTS_TAKEHOME.csv')

# 1. General overview of the dataset: number of rows, columns, and data types
def dataset_overview(df):
    print("Dataset Overview")
    print("Number of Rows: ", df.shape[0])
    print("Number of Columns: ", df.shape[1])
    print("\n Column Data Types")
    print(df.dtypes)
    print("\n")

# 2. Summary of missing values
def missing_values_summary(df):
    print("Missing Values Summary")
    missing_values = df.isnull().sum()
    missing_percentage = (df.isnull().sum() / df.shape[0]) * 100
    print(pd.DataFrame({'Missing Values': missing_values, 'Percentage': missing_percentage}))
    print("\n")

# Generate the dataset summary
def generate_summary(df):
    dataset_overview(df)
    missing_values_summary(df)

generate_summary(df)
```

## Visualizations

### Distribution of Products by Primary Category (CATEGORY_1)

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Count products by CATEGORY_1
category_counts = df['CATEGORY_1'].value_counts()

# Plot the distribution of products by CATEGORY_1
plt.figure(figsize=(12, 6))
sns.barplot(x=category_counts.index, y=category_counts.values)
plt.title('Distribution of Products by Primary Category (CATEGORY_1)')
plt.xticks(rotation=90)
plt.xlabel('Primary Category')
plt.ylabel('Number of Products')
plt.show()
```

### Top 10 Brands by Product Count

```python
# Count products by BRAND and get the top 10 brands
brand_counts = df['BRAND'].value_counts().head(10)

# Plot the top 10 brands by number of products
plt.figure(figsize=(12, 6))
sns.barplot(x=brand_counts.index, y=brand_counts.values)
plt.title('Top 10 Brands by Product Count')
plt.xticks(rotation=90)
plt.xlabel('Brand')
plt.ylabel('Number of Products')
plt.show()
```

### Top 10 Manufacturers by Product Count


```python
# Count products by MANUFACTURER and get the top 10 manufacturers
manufacturer_counts = df['MANUFACTURER'].value_counts().head(10)

# Plot the top 10 manufacturers by number of products
plt.figure(figsize=(12, 6))
sns.barplot(x=manufacturer_counts.index, y=manufacturer_counts.values)
plt.title('Top 10 Manufacturers by Product Count')
plt.xticks(rotation=90)
plt.xlabel('Manufacturer')
plt.ylabel('Number of Products')
plt.show()
```
